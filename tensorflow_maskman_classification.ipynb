```python
import numpy as np
import pandas as pd
import os
import glob
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras import Sequential, models
from tensorflow.keras.layers import Flatten, Dense, Conv2D, MaxPool2D
import matplotlib.pyplot as plt
import seaborn as sns
# from skimage.io import imread
# from skimage.transform import resize
import os
import sys
import shutil
```

```python
path = '/Users/jungjihoon/Desktop/maskman/Face Mask Dataset/'
dataset = {
    "image_path" : [],
    'mask_status': [],
    'where' : []
}

for where in os.listdir(path)[1:]:
    for status in os.listdir(path + '/' + where):
        for image in glob.glob(path + where +'/' + status + '/' + '*.png'):
            dataset['image_path'].append(image)
            dataset['mask_status'].append(status)
            dataset['where'].append(where)
            
dataset = pd.DataFrame(dataset)
dataset.head()
```

```python
train_df = dataset[dataset['where'] == 'Train']
test_df = dataset[dataset['where'] == 'Test']
val_df = dataset[dataset['where'] == 'Validation']

train_df.head()
```

```python
figs = (15,5)

plt.figure(figsize=figs)
plt.subplot(1,3,1)
sns.countplot(x=train_df['mask_status'])
plt.title('train dataset', size=10)

# plt.figure(figsize=figs)
plt.subplot(1,3,2)
sns.countplot(x=test_df['mask_status'])
plt.title('test dataset', size=10)

# plt.figure(figsize=figs)
plt.subplot(1,3,3)
sns.countplot(x=val_df['mask_status'])
plt.title('validatoin dataset', size=10)

plt.show()
```

```python
train_df = train_df.reset_index().drop('index', axis=1)
train_df
```

```python
# gray 처리 후 이미지 resize한 후 data에 저장
data = []
image_size = 150

for i in range(len(train_df)):
    ##converting the image into grayscale
    img_array = cv2.imread(train_df['image_path'][i], cv2.IMREAD_GRAYSCALE)
    
    # resizing the array
    new_image_array = cv2.resize(img_array, (image_size, image_size))
    
    # encoding the image with the label
    if train_df['mask_status'][i] == 'WithMask':
        data.append([new_image_array, 1])
    else:
        data.append([new_image_array, 0])
```

```python
data = np.array(data)
type(data)
```

```python
np.random.shuffle(data)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(10,6))

for row in range(2):
    for col in range(3):
        image_index = row*100 + col
        
        ax[row, col].axis('off')
        ax[row, col].imshow(data[image_index][0], cmap='gray')
        
        if data[image_index][1] == 0:
            ax[row, col].set_title('without mask')
            
        else:
            ax[row, col].set_title('with mask')
            
plt.show()
```

```python
X = []
y = []

for image in data:
    X.append(image[0])
    y.append(image[1])
    
X = np.array(X)
y = np.array(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=13)
X_train.shape, X_val.shape
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, kernel_size=(5, 5), strides=(1,1), padding='same', activation='relu', input_shape=(150,150,1)),
    layers.MaxPool2D(pool_size=(2,2), strides=(2,2)),
    layers.Conv2D(64, (2,2), activation='relu', padding='same'),
    layers.MaxPool2D(pool_size=(2,2)),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(1000, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam',
             loss=tf.keras.losses.BinaryCrossentropy(),
             metrics=['accuracy'])
```

```python
X_train = X_train.reshape(len(X_train), X_train.shape[1], X_train.shape[2], 1)
X_val = X_val.reshape(len(X_val), X_val.shape[1], X_val.shape[2], 1)
history = model.fit(X_train, y_train, epochs=5, batch_size=32)
```

```python
model.evaluate(X_val, y_val)
```

```python
prediction = model.predict(X_val)
```

```python
prediction = np.around(prediction,0).astype(int)
print("classification_report", classification_report(y_val, prediction))
print('confusion_matrix', confusion_matrix(y_val, prediction))
```

```python
import itertools

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize=True'.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks=np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# compute the confusion matrix
print(confusion_matrix(y_val, prediction))
confusion_mtx = confusion_matrix(y_val, prediction)
# plot the confusion matrix
plot_confusion_matrix(confusion_mtx, classes = range(2))
```
